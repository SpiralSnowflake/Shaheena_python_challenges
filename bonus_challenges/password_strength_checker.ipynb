Challenge: Password Strength Checker
Problem
1. Create a program that checks whether a password is strong based on certain criteria. A strong password might include uppercase letters, lowercase letters, numbers, and symbols.

Example
Enter a password: myPass123  
Result: Weak (missing symbol)
Your Task
1. Get a password from user input
2. Check for the following:
3. Minimum length (e.g., 8 characters)
4. Contains at least one uppercase letter
5. Contains at least one lowercase letter
6. Contains at least one number
7. Contains at least one symbol
8. Return a message that rates the password as “Strong,” “Medium,” or “Weak”
Concepts Tested
String methods
Conditionals and loops
User input
Stretch Ideas
2. Use regular expressions for advanced validation
3. Suggest how to improve the password
4. Allow repeated input until a strong password is given

```python
# Create a program that checks whether a password is strong based on certain criteria. A strong password might include uppercase letters, lowercase letters, numbers, and symbols.
# 1. Get a password from user input
import string
def password_strength_checker():
    user_input_password = input("Enter the password : ")
# 2. Check for the following: # 3. Minimum length (e.g., 8 characters) # 4. Contains at least one uppercase letter
# 5. Contains at least one lowercase letter# 6. Contains at least one number # 7. Contains at least one symbol
    length_count = False
    has_upper = False
    has_lower = False
    has_digit = False
    has_symbol = False
    password_strength_score = 0
    if len(user_input_password) >= 8:
        length_count = True
    for char in user_input_password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif char in string.punctuation:
            has_symbol = True
    
            
# 8. Return a message that rates the password as “Strong,” “Medium,” or “Weak” # IF strength_score == 5: "Very Strong Password"
# IF strength_score >= 3: "Moderate Password" if strength_score <3 "Weak Password"
    if length_count== True:
        password_strength_score += 1
    if has_upper == True:
        password_strength_score +=1
    if has_lower == True:
        password_strength_score +=1
    if has_digit == True:
        password_strength_score +=1
    if has_symbol == True:
        password_strength_score +=1

    
    if password_strength_score == 5:
        print(f"Strong Password.")
    elif password_strength_score >= 3:
        print(f"Medium Password.")
    else:
        print(f"Weak Password.")
    print(f"You entered {user_input_password}")
    if length_count== False:
        print("Your password length should be 8 characters or more." )  
    if has_upper == False:
        print("Your password is missing uppercase letter.")
    if has_lower == False:
        print("Your password is missing lowercase letter.")
    if has_digit == False:
        print("Your password is missing a number.")
    if has_symbol == False:
        print("Your password is missing a symbol.")
    return
        
print(password_strength_checker())
```
