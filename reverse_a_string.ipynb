Reverse a String
Challenge: Reverse That String!
1. Problem

Write a function that takes a single string as input and returns that string in reverse order.

Example
reverse_string("hello") # will output: olleh
reverse_string("Python") # will output: nohtyP
Your Task
Define a function called reverse_string that accepts one argument (a string).
Return the reversed version of the string.
Try using string slicing as a first approach.
2. Bonus: After you get it working, try writing a second version using a for loop to build the reversed string manually.

```python
# Reverse  string using Slicing 
def reverse_string_slicing(user_input): 
    rev_string = user_input[::-1]
    return(rev_string)
        
user_input = input("Enter a string : ")
print(f"Reversed String of {user_input} is {reverse_string_slicing(user_input)}")
    
```

```python
# Reverse  string using for loop 
def reverse_string_for(user_input: str):
    rev_string = ""
    for char in user_input:
        rev_string = char + rev_string
    return(rev_string)
         
user_input = input("Enter a string : ")
print(f"Reversed String of {user_input} is : {reverse_string_for(user_input)}")
```

```python
#Add a feature that ignores whitespace or punctuation at the beginning and end using strip() and string package # strip(string.punctuation)
import string
def reverse_string_for(user_input_s):
    rev_string = ""
    for char in user_input_s:
        rev_string = char + rev_string
    return(rev_string.strip())
         
user_input = input("Enter a string : ")
user_input_s = user_input.strip().strip(string.punctuation) #strip() method removes any leading or trailing whitespace characters (spaces, tabs, newlines, etc.) 
                                                            #strip(string.punctuation) removes punctuation (!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~) from the beginning and end of the string                       
print(f"Reversed String of {user_input} is : {reverse_string_for(user_input_s)}")
```

```python
# Make the function work for lists as well as strings
def reverse_string_slicing(user_input1: list):
    rev_list = user_input1[::-1]
    return(rev_list)
     
user_input1 =  input("Enter a list separated by commas : ")
print(f"Reversed List of {user_input1} :  {reverse_string_slicing(user_input1)}")
    
```
